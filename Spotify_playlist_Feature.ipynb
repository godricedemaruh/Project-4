```python
df = df.drop(columns={'mode', 'key', 'speechiness', 'liveness', 'time_signature', 'duration_ms'})
```

```python
def min_max_scaling(col):
    min_col = min(col)
    max_col = max(col)
    norm_val = []
    for val in col:
        norm_val.append((val - min_col) / (max_col - min_col))
    return norm_val

def range_scaling_0_1(col):
    min_col = 0
    max_col = 1
    norm_val = []
    for val in col:
        norm_val.append((val - min_col) / (max_col - min_col))
    return norm_val

def range_scaling_0_34(col):
    min_col = 0
    max_col = 34
    norm_val = []
    for val in col:
        norm_val.append((- val - min_col) / (max_col - min_col))
    return norm_val

# no range scaling for tempo; max and min values...
```
